```python
import math
def radians_to_degrees(radians):
    return radians * 180 / math.pi
```

```python
print(radians_to_degrees(3.14))
```

```python
def sort_numbers(numbers, order):
    if order == "asc":
        return sorted(numbers)
    elif order == "desc":
        return sorted(numbers, reverse=True)
    elif order == "none":
        return numbers
    else:
        return numbers
```

```python
print(sort_numbers([4, 2, 9, 1], "asc"))
print(sort_numbers([4, 2, 9, 1], "desc"))
```

```python
def decimal_to_binary(n):
    return bin(n)[2:]
```

```python
print(decimal_to_binary(10))
```

```python
def count_vowels(word):
    vowels = "aeiou"
    count = 0
    for char in word.lower():
        if char in vowels:
            count += 1
    return count
```

```python
print(count_vowels("hello"))
```

```python
def hide_credit_card(card_number):
    return "*" * (len(card_number) - 4) + card_number[-4:]
```

```python
print(hide_credit_card("4444444444444444"))
```

```python
def xo_equal(text):
    text = text.lower()
    return text.count("x") == text.count("o")
```

```python
print(xo_equal("xoxo"))
```

```python
def calculator(a, operator, b):
    if operator == "+":
        return a + b
    elif operator == "-":
        return a - b
    elif operator == "*":
        return a * b
    elif operator == "/":
        return a / b
```

```python
print(calculator(6, "*", 4))
```

```python
def apply_discount(price, discount):
    return price - (price * discount / 100)
```

```python
print(apply_discount(100, 20))
```

```python
def only_numbers(items):
    result = []
    for item in items:
        if isinstance(item, int):
            result.append(item)
    return result
```

```python
print(only_numbers([1, "a", 3, "hello", 5]))
```

```python
def double_characters(text):
    result = ""
    for char in text:
        result += char * 2
    return result
```

```python
print(double_characters("now"))
```

```python
def to_uppercase(text):
    return text.upper()
```

```python
print(to_uppercase("goodbye"))
```

```python
def add_dots(text):
    return ".".join(text)
```

```python
print(add_dots("skills"))
```
